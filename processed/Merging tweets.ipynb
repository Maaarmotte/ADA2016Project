```python
import csv
import glob
import os
import sys
import pandas as pd
```

```python
data_path = "processed/bayes"
outfile_path = "BayesCombined.csv"
filewriter = csv.writer(open(outfile_path,'wb'))
file_counter = 0
for input_file in glob.glob(os.path.join(data_path,'*.csv')):
    with open(input_file,'rU') as csv_file:
        filereader = csv.reader(csv_file)
        #name, ext = os.path.splitext(input_file)
        fn = input_file.split('_')
        month = fn[5]
        if file_counter < 1:
            for i, row in enumerate(filereader):
                if i==0: row.append('Month')
                else: row.append(month)
                filewriter.writerow(row)
            else:
                header = next(filereader,None)
                for row in filereader:
                    row.append(month)
                    filewriter.writerow(row)
    file_counter += 1
```
