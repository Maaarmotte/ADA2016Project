### Combining the produced .csv files into one file, with month and language columns.

```python
import csv
import glob
import os
import sys
import pandas as pd
```

Defining the input file and the output file and the column names for the new dataframe.

```python
inFile = "processed/Bayes"
outFile = "bayesCombined.csv"

os.chdir(inFile)
fileList=glob.glob("*.csv")
dfList=[]
colNames = ["City", "Sentiment", "Count", "Month", "Language"]
```

Extracting the language and month from the filenames. Concatenating the files into one dataframe and writing it to the output file.

```python
for filename in fileList:
    print(filename)
    fn = filename.split('_')
    lang=fn[3]
    fn= fn[4].split('.')
    month = fn[0]
    df=pd.read_csv(filename,header=None)
    df["Month"]=month
    df["Language"]=lang
    dfList.append(df)
    
concatDf=pd.concat(dfList,axis=0)
concatDf.columns=colNames
concatDf.to_csv(outFile,index=None)
    
```
